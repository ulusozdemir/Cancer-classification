```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

import lightgbm as lgb
```

```python
featureData = pd.read_csv("data.csv")
```

```python
featureData
```

```python
labelData = pd.read_csv("labels.csv")
```

```python
labelData
```

```python
features = featureData[featureData.columns[1:]].values
labels = labelData[labelData.columns[1]].values
```

```python
features
```

```python
labels
```

```python
featuresTrain, featuresTest, labelsTrain, labelsTest = train_test_split(features, labels, test_size=0.4, random_state=7)
```

```python
labelsTrain
```

```python
labelsTest
```

```python
# Create a Random Forest classifier with 100 trees
rf_classifier = RandomForestClassifier(n_estimators=100)

# Train the classifier
rf_classifier.fit(featuresTrain, labelsTrain)
```

```python
clf = lgb.LGBMClassifier()


clf.fit(featuresTrain, labelsTrain)
```

```python
# Predict the classes for the test dataset
predictedLabelsForest = rf_classifier.predict(featuresTest)
predictedLabelsForest
```

```python
predictedLabelsForest.shape
```

```python
predictedLabelsBoosted = clf.predict(featuresTest)
predictedLabelsBoosted
```

```python
predictedLabelsBoosted.shape
```

```python
# Calculate the confusion matrix
confusion_mat = confusion_matrix(labelsTest, predictedLabelsForest)
confusion_mat
```

```python
# Extract the values from the confusion matrix
true_positives = confusion_mat.diagonal()
false_negatives = confusion_mat.sum(axis=1) - true_positives
false_positives = confusion_mat.sum(axis=0) - true_positives
true_negatives = confusion_mat.sum() - (true_positives + false_negatives + false_positives)

# Calculate sensitivity and specificity
sensitivity = np.mean(true_positives / (true_positives + false_negatives))
specificity = np.mean(true_negatives / (true_negatives + false_positives))

accuracy = np.sum(predictedLabelsForest == labelsTest) / len(labelsTest)

# Print the sensitivity and specificity
print("Random Forest Performance Measures")
print("----------------------------------")
print("Sensitivity:", sensitivity)
print("Specificity:", specificity)
print("Accuracy:", accuracy)
```

```python
confusion_mat = confusion_matrix(labelsTest, predictedLabelsBoosted)
confusion_mat
```
